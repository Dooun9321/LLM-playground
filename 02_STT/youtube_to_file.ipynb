유튜브 쇼츠를 파일로 다운받기

```python
import yt_dlp

# url = "https://www.youtube.com/shorts/JLszYpIAj0Q"  # 쇼츠 URL
url = "https://www.youtube.com/shorts/L7V6DgmRFFY"

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    # 'outtmpl': 'audio.%(ext)s',
    'outtmpl': 'video.%(ext)s',
    'ffmpeg_location': '/opt/homebrew/bin/ffmpeg',  # ffmpeg 경로 명시
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("다운로드 완료! ✅")
```
